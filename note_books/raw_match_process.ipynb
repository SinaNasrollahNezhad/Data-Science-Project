```python
import polars as pl
from glob import glob
import os
```

```python
days_address = r"G:\learning\daneshkar\statistic and data analis\project\main\data\unzip_data"
out_put_folder = r"G:\learning\daneshkar\statistic and data analis\project\Data-Science-Project\data\proccessed_data\raw_match"
os.makedirs(out_put_folder,exist_ok=True)


day_names = [name for name in os.listdir(days_address) if len(name) == 8 and name.isdigit()]
day_names.sort()
print(f"numbers of date is {len(day_names)}")


match_types = [
    "away_team",
    "away_team_score",
    "event",
    "home_team",
    "home_team_score",
    "round",
    "season",
    "time",
    "tournament",
    "venue"
    
]

print(f"Processing raw_match_parquet from {len(day_names)} days...\n")
```

```python
for m_type in match_types:
    all_combined = pl.DataFrame()  
    
    print(f"--- Processing type: {m_type} ---")
    
    for day in day_names:
        base = f"{days_address}/{day}/data/raw/raw_match_parquet"
        
        
        pattern = f"{base}/{m_type}*.parquet"
        files = glob(pattern)
        
        if not files:
            continue
        
        print(f"   Day {day}: {len(files)} files")
        
        dfs = [pl.read_parquet(f) for f in files]
        day_combined = pl.concat(dfs, how='diagonal_relaxed')
        
       
        all_combined = pl.concat([all_combined, day_combined], how='diagonal_relaxed')
    
   
    if not all_combined.is_empty():
        output_file = f"{out_put_folder}/{m_type}_all_days.parquet"
        all_combined.write_parquet(output_file)
        print(f"Saved: {output_file} | Shape: {all_combined.shape}\n")
    else:
        print(f"No data for {m_type}\n")

print("Strategy 2 (raw_match_parquet split) completed")
```
