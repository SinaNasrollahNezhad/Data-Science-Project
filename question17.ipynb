```python
import pandas as pd
```

```python
df_games = pd.read_parquet(r'D:\daneshkar\pandas-numpy-matplotlib\x\final_datasets\raw_tennis_power_parquet_all_days.parquet')
df_games
```

```python
df_stats = pd.read_parquet(r'D:\daneshkar\pandas-numpy-matplotlib\x\final_datasets\raw_statistics_parquet_all_days.parquet')
df_stats
```

```python
break_per_match = df_games.groupby('match_id')['break_occurred'].sum().reset_index()
break_per_match.rename(columns={'break_occurred': 'total_breaks'}, inplace=True)
break_per_match
```

```python
df_serves = df_stats[
    (df_stats['statistic_category_name'] == 'service') & 
    (df_stats['statistic_name'] == 'first_serve')
].copy()
filter_serves = df_serves[['match_id' , 'statistic_category_name'  , 'statistic_name']]
filter_serves
```

```python
merged_total = pd.merge( filter_serves , break_per_match , on= 'match_id' )
merged_total
```

```python
merged_total.groupby('match_id')['total_breaks'].mean()
```
