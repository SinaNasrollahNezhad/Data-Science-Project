```python
import pandas as pd
import polars as pl
from glob import glob
import os
```

```python
days_address = r"G:\learning\daneshkar\statistic and data analis\project\main\data\unzip_data"
out_put_folder = r"G:\learning\daneshkar\statistic and data analis\project\Data-Science-Project\data\proccessed_data"
os.makedirs(out_put_folder,exist_ok=True)

folders = [
    "raw_odds_parquet",
    "raw_point_by_point_parquet",
    "raw_statistics_parquet",
    "raw_tennis_power_parquet",
    "raw_votes_parquet"]



day_names = [name for name in os.listdir(days_address) if len(name) == 8 and name.isdigit()]
day_names.sort()
print(f"numbers of date is {len(day_names)}")
```

```python

for folder in folders:
    all_combined = pl.DataFrame()  
    
    print(f"--- Processing folder: {folder} ---")
    
    for day in day_names:
        base = f"{days_address}/{day}/data/raw/{folder}"
        
        files = glob(f"{base}/*.parquet")
        
        if not files:
            continue
        
        print(f"   Day {day}: {len(files)} files")
        
        dfs = [pl.read_parquet(f) for f in files]
        
        if dfs:
            day_combined = pl.concat(dfs, how='diagonal_relaxed')
            all_combined = pl.concat([all_combined, day_combined], how='diagonal_relaxed')
    

    if not all_combined.is_empty():
        output_file = f"{out_put_folder}/{folder}_all_days.parquet"
        all_combined.write_parquet(output_file)
        print(f"Saved: {output_file} | Shape: {all_combined.shape}\n")

print("Processing complete")
```
